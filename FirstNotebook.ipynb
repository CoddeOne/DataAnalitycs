```python
def task_10(N):
    found = False
    for a in range(1, N):
        for b in range(a, N):
            if a**2 + b**2 == N:
                print(f'{a}^2 + {b}^2 = {N}')
                found = True
                break
        if found:
            break
    if not found:
        print(f'Число {N} не можна представити квадратами цілих чисел')

# Приклади
print("Приклад 1:")
task_10(45)
print("\nПриклад 2:")
task_10(21)
```
